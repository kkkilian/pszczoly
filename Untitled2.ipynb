```python
import tensorflow as tf
import tensorflow_datasets as tfds
import matplotlib.pyplot as plt
import numpy as np
```

```python
dataset = tfds.load('bee_dataset', split=['train'], as_supervised=True)

array = np.vstack(tfds.as_numpy(dataset[0]))

X_train = np.array(list(map(lambda x: x[0], array)))
y_train = np.array(list(map(lambda x: x[1], array)))
```

```python
def plot(X, ile):
  fig = plt.figure(figsize= (22, 22))
  for i in range (0,ile):
    ax = fig.add_subplot(ile/2, ile/2, i+1)
    ax.imshow(X[i])
```

```python
plot(X_train, 16)
```
